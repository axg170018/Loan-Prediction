# SPLITTING THE DATA INTO TRAINING AND TESTING DATA

```python
from sklearn.model_selection import train_test_split
X_train_traindata, X_test, Y_train_traindata, y_test = train_test_split( X_train, Y_train, test_size=0.2, random_state=4)
print ('Train set:', X_train_traindata.shape,  Y_train_traindata.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k = 3
#Train Model and Predict  
neigh = KNeighborsClassifier(n_neighbors = k).fit(X_train_traindata,Y_train_traindata)
neigh
```

```python
yhat = neigh.predict(X_test)
yhat[0:5]
```

```python
yhat
```

```python
from sklearn import metrics
print("Train set Accuracy: ", metrics.accuracy_score(Y_train_traindata, neigh.predict(X_train_traindata)))
print("Test set Accuracy: ", metrics.accuracy_score(y_test, yhat))
```
